```python
comments = df["Comments"]
print(comments)
```

```python
def merge_users(comments):
    users_list = []
    for com in comments:
         users_list.extend(list(com.keys()))
    return users_list

users_list = merge_users(comments)

```

```python
print(len(users_list))
print(users_list[:10])
```

```python
# Counts
```

```python
def create_counts(users_list):

    counts = {}
    for u in users_list:
        counts[u] = []

    for u in users_list:
        counts[u].append(u)
        
    for k,v in counts.items():
        counts[k] = len(v)
    
    sorted_twitts =  sorted(counts.items(), key= lambda kv:(kv[1], kv[0]))[::-1][:25]

    return sorted_twitts
counts = create_counts(users_list)
```

```python
counts
```

```python
def create_df(counts):

    df_x = pd.DataFrame(counts, columns=["label", "count"])
    df_x["index"] = df_x.index

    df_x.index = list(df_x["label"])

    df_x = df_x.sort_values(['count'], ascending=False)

    return df_x

df_counts = create_df(counts)
```

```python
def plot_word_count(df, title):
    plt.figure(figsize=(13,8))
    ax = sns.barplot(x="index", y="count", data=df, order=df['index'])
    ax.set_xlabel(title)
    ax.set_ylabel('Number of Comments')
    ax.set_xticklabels(df['label'], rotation='vertical', fontsize=10) 
```

```python
top20 = df_counts.iloc[0:20]

plot_word_count(top20, f"Top {len(top20)} People")
```
